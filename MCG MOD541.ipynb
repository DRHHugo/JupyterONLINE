# Este programa define un MCG con mo√≥ulo 541. El multiplicador y la semilla son ingresados como argumentos

```python
def MCGmod541(b,x0,n):
    sample=[b*x0%541]
    for i in range(n-1):
        sample.append(b*sample[-1]%541)
    return sample
```
